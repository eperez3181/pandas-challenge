```python
# Dependencies and Setup
import pandas as pd

# File to Load 
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# DISTRICT SUMMARY *************************************************************************************************
school_data_complete.head()
```

```python
school_data_df = school_data
school_data_df.head()
```

```python
total_schools = len(school_data_df["school_name"].unique())
total_schools
```

```python
total_students = len(student_data["Student ID"].unique())
total_students
```

```python
total_budget =school_data_df["budget"].sum()
total_budget
```

```python
avg_math_score = school_data_complete["math_score"].mean()
avg_math_score
```

```python
avg_reading_score = school_data_complete["reading_score"].mean()
avg_reading_score
```

```python
passing_math_percentage = (len(school_data_complete.loc[school_data_complete["math_score"] >= 70]) / total_students )*100
passing_math_percentage
```

```python
passing_reading_percentage = (len(school_data_complete.loc[school_data_complete["reading_score"] >= 70]) / total_students )*100
passing_reading_percentage
```

```python
overall_passing = school_data_complete.loc[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"]>=70)]
```

```python
overall_passing_percent_df =(len(overall_passing) / total_students)*100
overall_passing_percent_df
```

```python
# Create a DataFrame for District Summary that contains 8 columns 
# and has 4 rows of data
district_summary_df = pd.DataFrame(
    {"Total Schools": [total_schools],
     "Total Students": [total_students],
      "Total Budget": [total_budget],
      "Average Math Score": [avg_math_score],
      "Average Reading Score":[avg_reading_score],
      "% Passing Math":[passing_math_percentage],
      "% Passing Reading":[passing_reading_percentage],
      "% Overall Passing":[overall_passing_percent_df]})

#district_summary_df 
```

```python
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df
```

```python
# SCORES BY SCHOOL SPENDING ***************************************************************************************
# Figure out the minimum and maximum Per Student Budget
print(new_school_summary_df["Per Student Budget"].max())
print(new_school_summary_df["Per Student Budget"].min())
```

```python
# Create bins in which to place values based upon Per Student Budget
bins = [0, 584, 629, 644, 680]

# Create labels for the bins
group_labels = ["<$585", "585-630", "630-645", "645-680"]
```

```python
# Slice the data and place it into bins
pd.cut(new_school_summary_df["Per Student Budget"], bins, labels=group_labels)
```

```python
# Place the data series into a new column inside of the DataFrame
new_school_summary_df["Spending Ranges (Per Student)"] = pd.cut(new_school_summary_df["Per Student Budget"], bins, labels=group_labels)
new_school_summary_df.head()
```

```python
# Create a GroupBy object based on "Spending Ranges (Per Student)"
spending_ranges = new_school_summary_df.groupby("Spending Ranges (Per Student)")

# Find how many rows fall into each bin
print(spending_ranges["Per Student Budget"].count())

# Get the average of the needed columns
scores_by_school_spending = spending_ranges[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]].mean()
#scores_by_school_spending
```

```python
scores_by_school_spending["Average Math Score"] = scores_by_school_spending["Average Math Score"].map("{:.2f}".format)
scores_by_school_spending["Average Reading Score"] = scores_by_school_spending["Average Reading Score"].map("{:.2f}".format)
scores_by_school_spending["% Passing Math"] = scores_by_school_spending["% Passing Math"].map("{:.2f}".format)
scores_by_school_spending["% Passing Reading"] = scores_by_school_spending["% Passing Reading"].map("{:.2f}".format)
scores_by_school_spending["% Overall Passing"] = scores_by_school_spending["% Overall Passing"].map("{:.2f}".format)
scores_by_school_spending
```
